```python
#To plot |Y|^2

import special_functions as sf
import numpy as np
import matplotlib.pyplot as plt
import math as mt

```

```python
thetas = np.linspace(0,mt.pi,100)
phis = np.linspace(0,2*mt.pi,100)
theta, phi = np.meshgrid(thetas,phis)
#custom value of l,m
Ylm1 = sf.Ylm(2,2,theta,phi)
R = np.real(np.conjugate(Ylm1)*Ylm1)

X, Y, Z = R*np.cos(phi)*np.sin(theta), R*np.sin(phi)*np.sin(theta), R*np.cos(theta)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X,Y,Z, cmap = 'winter')
```
