```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
C = -0.1
D = -0.0225
s = 0.5

def skey(ene):
    return ene[0]
Energy=[]
l_char = ['s','p','d','f','g','h','i','j','k','l']
for N in range(7+1):
    for l in range(N,-1,-2):
        n = int((N-l)*0.5+1)
        if l!=-1:
            
            j=l+s
            E = (N + 1.5) + C*0.5*l + D*l*(l+1)
            Energy.append([round(E,4),N, N+1.5, n,l_char[l],j,2*j+1])
            if l!=0:
                j = l-s
                E = (N + 1.5) - C*0.5*(l+1) + D*l*(l+1)
                Energy.append([round(E,4),N, N+1.5, n,l_char[l],j,2*j+1])

Energy.sort(key=skey)
```

```python
fig = plt.figure(figsize=(15,75))
ax = fig.add_subplot()
ax.set_xlim([-2, 36])
ax.set_ylim([1.45, 7.1])
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
plt.title('Single-particle energy level positions for the spherical shell model', fontsize = 20)
plt.ylabel('Energy', fontsize = 30)

x_values = [0, 7.5, 12.5, 30]

for i in range(len(Energy)):
    N = Energy[i][1]
    if N<6:
        plt.text(2.5,Energy[i][2]+0.01,'N =', fontsize = 15)
        plt.text(5.2,Energy[i][2]+0.01, N , fontsize = 15)
    
    if Energy[i][0]<7:
        plt.text(31,Energy[i][0], Energy[i][3], fontsize = 13)
        plt.text(31.7,Energy[i][0], Energy[i][4], fontsize = 13)
        plt.text(32.7,Energy[i][0], int(2*Energy[i][5]), fontsize = 13)
        plt.text(33.9,Energy[i][0], '/', fontsize = 13)
        plt.text(34.5,Energy[i][0], '2', fontsize = 13)
        plt.text(25,Energy[i][0]+0.003, int(Energy[i][6]), fontsize = 10)
        
    y_values = [Energy[i][2], Energy[i][2], Energy[i][0], Energy[i][0]]
    plt.plot(x_values, y_values)
```
